```python
# tutorial from: https://www.youtube.com/watch?v=iOVXRpW9BX0 #

import numpy as np
from matplotlib import pyplot

'''Geometry'''
s_B = 0.5   # meters

'''Spatial Grid'''
dx = 0.01   # meters
nx = int(s_B/dx)
x = np.linspace(0,s_B,nx)

'''Time Grid'''
t_sim = 300   # seconds
dt = 0.5   # seconds
nt = int(t_sim/dt)

'''Material Properties'''
rho = 2000   # kg/m^3
cp = 500   # J/kg/K
lamda = 10   # W/m/K

'''Initial Condition'''
T_A = 303.15   # K

'''Define Results Parameter'''
T = np.ones(nx)*T_A

'''Simulation'''
for n in range(1,nt):
    
    Tn = T.copy()
    
    T[1:-1] = Tn[1:-1] + dt*lamda/(rho*cp) * (Tn[2:] - 2*Tn[1:-1] + Tn[0:-2])/dx**2
    
    T[0] = 293.15   # K
    T[-1] = T[-2]   # K
    
plotT = pyplot.figure(figsize=(6,7), dpi=100)
pyplot.plot(x,T-273.15,'k')
pyplot.show(plotT)
```
